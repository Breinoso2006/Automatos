```python
#Classe Quintupla
class Quintupla:
    def __init__(self, estadoAtual, leitura, escrita, proxEstado, direcao):
        self.estadoAtual = estadoAtual
        self.leitura = leitura
        self.escrita = escrita
        self.proxEstado = proxEstado
        self.direcao = direcao

#Máquina de Turing
def TuringMachine(fita, listaDeQuintuplas):
    print('Fita Inicial:', fita,'\n')
    estadoAtual = 0
    posiFita = 0
    aux = 0
    listaEstados = []
    listaLeitura = []
    
    #Faz uma lista com os estados existentes 
    for quint in listaDeQuintuplas:
        listaEstados.append(listaDeQuintuplas[aux].estadoAtual)
        aux +=1
    aux = 0
    
    #Faz uma lista com as leituras possíveis
    for quint in listaDeQuintuplas:
        listaLeitura.append(listaDeQuintuplas[aux].leitura)
        aux +=1
    aux = 0
    
    #Define a posição inicial do cabeçote
    while (fita[posiFita] == 'b'):
        posiFita += 1        
    
    #Máquina de Turing
    print('Entrou na máquina.')
    while (estadoAtual in listaEstados and posiFita >= 0 and posiFita < len(fita)):
        if (aux >= len(listaDeQuintuplas) or listaDeQuintuplas[aux].leitura not in listaLeitura):
            print('Quebrou/Parou.')
            break
                        
        elif (estadoAtual == listaDeQuintuplas[aux].estadoAtual and \
              fita[posiFita] == listaDeQuintuplas[aux].leitura):
            fita[posiFita] = listaDeQuintuplas[aux].escrita
            estadoAtual = listaDeQuintuplas[aux].proxEstado
            print('Estou no estado', estadoAtual,'. Na posicao',posiFita,\
                  'da fita eu li',listaDeQuintuplas[aux].leitura, 'e escrevi ',\
                  listaDeQuintuplas[aux].escrita)
            posiFita += 1 if (listaDeQuintuplas[aux].direcao == 'R') else -1
            aux = -1 
        aux += 1
    print ("Saiu da Máquina.\n")
    print('Fita Final:', fita,'\n')    
```
