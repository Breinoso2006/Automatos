```python
#Classe Estado
class Estado:
    def __init__(self, numero, saida0, saida1, output):
        self.numero = numero
        self.saida0 = saida0
        self.saida1 = saida1
        self.output = output

#Máquina de Estados Finitos
def MEF(palavra, listaDeEstados):
    print('Palavra lida:', palavra)
    morto = 0
    simboloEstranho = 0
    
    for n in palavra:
        if n != 0 and n != 1:
            print("Esta máquina aceita apenas 0's e 1's.\n")
            simboloEstranho = 1
    
    if simboloEstranho == 0:
        #Definir o primeiro estado
        n = 0
        while n < len(listaDeEstados):
            if listaDeEstados[n].numero == 0:
                estadoAtual = listaDeEstados[n]
                break
            n+=1
        print('\nEstado inicial =', estadoAtual.numero)

        #Troca de Estados/Máquina
        while (len(palavra) != 0):
            aux = palavra[0]
            print("\nInput:",palavra[0])
            palavra.pop(0)

            if (aux == 0):
                if (estadoAtual.saida0 != 'morto'):
                    prox = estadoAtual.saida0
                else:
                    morto = 1
                    print('Não há estado para esta requisição.')
            elif (aux == 1):
                if (estadoAtual.saida1 != 'morto'):
                    prox = estadoAtual.saida1
                else:
                    morto = 1
                    print('Não há estado para esta requisição.')

            if (morto != 1):
                n = 0
                i = 0
                while (n < len(listaDeEstados)):
                    if (listaDeEstados[n].numero == prox):
                        estadoAtual = listaDeEstados[n]
                        break
                    n += 1
                print('Estado atual =' ,estadoAtual.numero, '\nOutput =', estadoAtual.output)
                i +=1
            else:
                break
    
        if (morto == 1):
            print('\nA máquina parou em um estado onde não pode concluir a requisição dada.\n')    
        elif (estadoAtual.output == 1):
            print('\nEssa palavra é reconhecida pela máquina!\n')
        else:
            print('\nEssa palavra não é reconhecida pela máquina!\n')
                
```

```python
#Chamada Main
if __name__ == '__main__':
    #Definir String
    palavra = [1,1,1,0,0,0,1,0,0]
    palavra2 = [1,1,1,0,0,0,1,1,0]
               
    #Definir Estados da Máquina
    listaDeEstados = [
        Estado(0,1,0,0), 
        Estado(1,2,0,0), 
        Estado(2,2,0,1)
    ]
    listaDeEstados2 = [
        Estado(0,1,0,0), 
        Estado(1,2,0,0), 
        Estado(2,2,'morto',1)
    ]
    
    #Chamar a Máquina
    MEF(palavra,listaDeEstados)
    MEF(palavra2, listaDeEstados2)
```
